```python
Data=pd.concat([D0220,D0320,D0420,D0520,D0620,D0720,D0820,D0920,D1020,D1120,D1220,D0121])
Data=D0121[{'Open time','Close'}]
Data.set_index('Open time',inplace=True)
Data
```

```python
train=Data.iloc[:200]
#val=Data.loc[datetime(2021,6,1) : datetime(2021,7,31)]
test=Data.iloc[200:]
```

```python
print(train.shape)
print(test.shape)
```

```python
# Feature Scaling
fs = MinMaxScaler(feature_range = (0, 1))
training_set_scaled = fs.fit_transform(train)
```

```python
# Creating a data structure with 30 time-steps (30 time period window) and 1 output
X_train = []
y_train = []
for i in range(20,200):
    X_train.append(training_set_scaled[i-20:i])
    y_train.append(training_set_scaled[i,0])
X_train, y_train = np.array(X_train), np.array(y_train)
```

```python
model = Sequential()

#Adding the first LSTM layer and some Dropout regularisation
model.add(LSTM(units = 100, return_sequences = True, input_shape = (X_train.shape[1], 1)))
model.add(Dropout(0.25))

# Adding a second LSTM layer and some Dropout regularisation
model.add(LSTM(units = 100, return_sequences = True))
model.add(Dropout(0.25))

# Adding a third LSTM layer and some Dropout regularisation
model.add(LSTM(units = 100, return_sequences = True))
model.add(Dropout(0.25))

# Adding a fourth LSTM layer and some Dropout regularisation
model.add(LSTM(units = 100))
model.add(Dropout(0.25))

# Adding the output layer
model.add(Dense(units = 1))

# Compiling the LSTM
model.compile(optimizer = 'adam', loss = 'mean_squared_error')

# Fitting the LSTM to the Training set
history=model.fit(X_train, y_train, epochs = 200)
```

```python
plt.figure(figsize=(14,8))
plt.plot(history.history['loss'])
plt.show()
```

```python
inputs = Data[len(Data.values)-len(test.values)-20:].values
print(inputs.shape)
inputs = inputs.reshape(-1,1)
inputs = fs.transform(inputs)
X_test = []
for i in range(20, len(inputs)):
    X_test.append(inputs[i-20:i])
X_test = np.array(X_test)
print(X_test.shape)
```

```python
preds=model.predict(X_test)
preds=fs.inverse_transform(preds)
```
